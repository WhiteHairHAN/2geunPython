```python
#  2019  2
#     :
# x(t) = (1+ 0.5cos(2*pi*t)) * cos(10*pi*t)
# (a)  x(t), -1 <=t <=1 [sec] 20000     .
from math import *
from matplotlib import pyplot as plt
import numpy as np

#x = np.arange(-1,1,0.0001)
x = np.linspace(-1,1,20000)

#x2 = np.arange(-1,1,0.0232)
y1 = (1+ 0.5*np.cos(2*pi*x)) * np.cos(10*pi*x)

#y2 = np.cos(4*np.pi*x)*np.exp(-2*x)

plt.plot(x,y1,'.',label ='(1+ 0.5*cos(2*pi*x)) * cos(10*pi*x)')
#plt.plot(x,y2,'o')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Telecommunication-Eng lecture 2019 mid 2nd problem')
plt.rcParams["figure.figsize"] = (20,10)
plt.legend(loc='upper right')
plt.show()
```
