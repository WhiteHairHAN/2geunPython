```python
#chapter.11
#11-6. tkinter를 사용하여 다음과 같은 사용자 인터페이스를 만든다.
# [출처] [두근두근 파이썬 풀이] CHAPTER 11|작성자 게으른 프로그래머 참고
#이런걸 시험에 낼까?

from tkinter import *
import pickle
import os

window = Tk()
window.title("tk")

button_list = ['추가', '처음', '다음','이전','마지막','파일 읽기']

nlabel = Label(window, width =7, text='이름')
nlabel.grid(row=0, column=0,columnspan = 2)

plabel = Label(window, width = 7, text='전화번호')
plabel.grid(row=1, column=0,columnspan = 2)

e1 = Entry(window, width=20)
e1.grid(row=0, column=2, columnspan = 4)

e2 = Entry(window, width=20)
e2.grid(row=1, column=2, columnspan = 4)

phone_book = {} #이름과 전화번호를 저장할 딕셔너리
phone_list = [] #전화번호를 여러개 등록할 경우를 위해 키의 값을 리스트로 저장
now_key = 0 #현재 키의 위치

#파일경로 예시: C:\Users\홍길동\Desktop\파이썬\두근두근파이썬 풀이\phone_book.dat
#r"파일경로" 앞에 r을 적어주면 \를 한번만 적어도 경로를 파악한다.
#r없이 파일경로를 입력하고 싶을 경우 아래와 같이 \\를 한다. 
#C:\\Users\\홍길동\\Desktop\\파이썬\\두근두근파이썬 풀이\\phone_book.dat

#이전에 파일이 존재한다면 해당 파일의 내용을 불러서 딕셔너리에 저장
if os.path.exists(r"파일경로\phone_book.dat"):
    outfile = open(r"파일경로\phone_book.dat", "rb")
    phone_book = pickle.load(outfile)
    outfile.close()

def click(key):
    global phone_book, phone_list, now_key
    if key == '추가':
        if e1.get() == "" or e2.get() == "": #만약 이름, 전화번호의 입력이 없다면 수행하지 않음
            pass
        elif e1.get() not in phone_book.keys():  # 새로운 이름이 입력되면
            phone_list.append(e2.get()) #리스트에 전화번호를 저장
            phone_book[e1.get()] = phone_list  # 새로운 키와 값을 딕셔너리에 저장시킴
            phone_list = []
        else:  # 이전에 존재하던 이름이라면
            phone_list = list(phone_book[e1.get()]) #키에 해당하는 값을 리스트로 변환
            if e2.get() in phone_list: #만약 키의 값을 저장한 리스트 내부에 동일한 값이 존재한다면
                e2.delete(0, END) #재입력을 위해 전화번호의 내용을 지움
            else:
                phone_list.append(e2.get())  # 리스트에 새로운 값을 추가
                phone_book[e1.get()] = phone_list  # 해당 리스트를 키의 값으로 변경
                phone_list = []
                
    elif key == '처음':
        e1.delete(0, END)
        e2.delete(0, END)
        now_key = 0
        if not phone_book: #딕셔너리에 값이 없다면
            e1.insert(0, "저장내역 없음")
            e2.insert(0, "저장내역 없음")
        else:
            phone_key = list(phone_book.keys())
            phone_value = phone_book[phone_key[0]] #키에 대한 값을 반환
            e1.insert(0, phone_key[0])
            e2.insert(0, phone_value)
            
    elif key == '다음':
        e1.delete(0, END)
        e2.delete(0, END)
        phone_key = list(phone_book.keys())
        if not phone_book:
            e1.insert(0, "저장내역 없음")
            e2.insert(0, "저장내역 없음")
        elif now_key >= (len(phone_key)-1): #키의 리스트 중 마지막과 키의 위치가 같다면
            e1.insert(0, "파일의 마지막.")
            e2.insert(0, "파일의 마지막.")
            now_key = (len(phone_key)) # 키의 위치를 1증가
        else:
            now_key += 1  # 키의 위치를 1증가
            phone_key = list(phone_book.keys())
            phone_value = phone_book[phone_key[now_key]]
            e1.insert(0, phone_key[now_key])
            e2.insert(0, phone_value)

    elif key == '이전':
        e1.delete(0, END)
        e2.delete(0, END)
        if not phone_book:
            e1.insert(0, "저장내역 없음")
            e2.insert(0, "저장내역 없음")
        elif now_key <= 0: #키의 처음위치 0보다 작을 경우
            e1.insert(0, "파일의 처음.")
            e2.insert(0, "파일의 처음.")
            now_key = 0
        else:
            now_key -= 1  # 키의 값을 1감소
            phone_key = list(phone_book.keys())
            phone_value = phone_book[phone_key[now_key]]
            e1.insert(0, phone_key[now_key])
            e2.insert(0, phone_value)
            
    elif key == '마지막':
        e1.delete(0, END)
        e2.delete(0, END)
        if not phone_book:
            e1.insert(0, "저장내역 없음")
            e2.insert(0, "저장내역 없음")
        else:
            e1.delete(0, END)
            e2.delete(0, END)
            phone_key = list(phone_book.keys())
            now_key = (len(phone_key) - 1) #키의 마지막으로 이동
            phone_value = phone_book[phone_key[now_key]]
            e1.insert(0, phone_key[now_key])
            e2.insert(0, phone_value)

    elif key == '파일 읽기':
        e1.delete(0, END)
        e2.delete(0, END)
        # 파일을 쓰기모드로 열기
        file = open(r"파일경로\phone_book.dat", 'wb')
        pickle.dump(phone_book, file)  # 딕셔너리가 저장하던 키와 값을 모두 파일에 작성
        file.close()  # 파일을 닫음과 동시에 저장됨
        # 파일을 읽기모으로 열기
        rfile = open(r"파일경로\phone_book.dat", 'rb')
        obj = pickle.load(rfile)  # 파일의 내용을 모두 읽어 옴
        print(obj)

num = 0
for i in button_list:
    def process(t = i):
        click(t)
    Button(window, text = i, command = process).grid(row = 3, column = num)
    num += 1

window.mainloop()
# [출처] [두근두근 파이썬 풀이] CHAPTER 11|작성자 게으른 프로그래머 참고
```
